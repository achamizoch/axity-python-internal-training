# Importar librerias

```python
import pandas as pd
import qgrid  as qg
from datetime import datetime
```

# Load Data

```python
df =  pd.read_csv('time_series_covid19_deaths_global.csv')
df.head(5)
```

```python
df.info()
```

# Clean data

```python
# Renombrar columnas
df.columns = list(df.columns[:4]) + [datetime.strptime(d, "%m/%d/%y").date().strftime("%Y-%m-%d") for d in df.columns[4:]]
df.rename(columns={"Province/State": "Province_State", "Country/Region": "Country_Region"}, inplace=True)
#Pivotear columnas de fechas a filas
df = df.melt(id_vars=['Country_Region', \
                      'Province_State', \
                      'Lat',            \
                      'Long'],          \
            value_vars=df.columns[4:],  \
            var_name  ='Date',          \
            value_name='Deaths')

df
```

# Process data

```python
df_agg = df.groupby('Country_Region')['Deaths'].sum().reset_index()
```

```python
df_agg1 = qg.show_grid(df_agg)
```
